```python
import pandas as pd
import re

def parse_price(price_str):
    """
    Attempts to convert price strings to a complete numeric value.
    
    Recognized formats:
      - '75L', '75 L', '75 lakh', '75lakh', '75 lac'  -> returns 75 * 100,000 = 7,500,000
      - '1.2Cr', '1.2 Cr', '1.2 crore'                 -> returns 1.2 * 10,000,000 = 12,000,000
      - 'Rs 75L', 'INR 1.2Cr'                          -> similar logic after removing currency symbols
      - If no unit is detected, returns the numeric part as-is.
    
    Unrecognized formats return None.
    """
    if not isinstance(price_str, str):
        return None
    
    # Convert to lowercase for uniformity
    price_str = price_str.lower().strip()
    
    # Remove common currency symbols or words (e.g., 'rs', 'inr')
    price_str = re.sub(r'\b(rs\.?|inr)\b', '', price_str)
    
    # Remove text within parentheses (e.g., "(negotiable)")
    price_str = re.sub(r'\(.*?\)', '', price_str)
    
    # Remove extra words that could interfere (e.g., 'onwards', 'approx')
    price_str = re.sub(r'\b(onwards|approx)\b', '', price_str)
    
    # Remove commas and extra spaces
    price_str = price_str.replace(',', '').strip()
    
    # Use regex to extract the first numeric part and an optional unit word.
    # This matches a number (possibly with commas or a decimal point) and
    # then an optional alphabetic unit.
    match = re.search(r'(\d[\d\.]*)\s*([a-z]+)?', price_str)
    if not match:
        return None

    numeric_part = match.group(1)
    try:
        value = float(numeric_part)
    except ValueError:
        return None

    unit = match.group(2) or ''

    # Check for unit keywords (for lakhs/lacs)
    if 'lakh' in unit or 'lac' in unit or unit == 'l':
        return value * 1e5  # 1 Lakh = 100,000
    # Check for unit keywords for crores
    elif 'cr' in unit or 'crore' in unit:
        return value * 1e7  # 1 Crore = 10,000,000
    else:
        # If no recognizable unit, assume the number is already complete.
        return value

# -------------------------------
# Example usage on your CSV file:
# -------------------------------

# Create a new column 'price_parsed' with the parsed values
df["price_parsed"] = df["price"].apply(parse_price)

# Print the number of rows where the price could not be parsed
print("Number of NaN (unparsed) prices:", df["price_parsed"].isna().sum())

# Save the updated DataFrame to a new CSV file
df.to_csv("output_parsed.csv", index=False)

print("Prices have been parsed and saved to 'output_parsed.csv'.")
```

```python
import pandas as pd
import re

def parse_price(price_str):
    """
    Parses price strings by first removing the ₹ symbol and converting values like:
      - '76.5 Lac' or '76.5lakh' -> 7,650,000
      - '1.13 Cr' or '1.13crore' -> 11,300,000
    into their complete numeric values.
    """
    if not isinstance(price_str, str):
        return None

    # Remove the rupee symbol
    price_str = price_str.replace("₹", "")
    
    # Convert to lowercase and strip whitespace
    price_str = price_str.lower().strip()
    
    # Remove common currency words if needed
    price_str = re.sub(r'\b(rs\.?|inr)\b', '', price_str)
    
    # Remove commas and extra spaces
    price_str = price_str.replace(',', '').strip()
    
    # Use regex to extract the numeric value and an optional unit
    match = re.search(r'(\d+(?:\.\d+)?)\s*([a-z]+)?', price_str)
    if not match:
        return None

    numeric_part = match.group(1)
    try:
        value = float(numeric_part)
    except ValueError:
        return None

    unit = match.group(2) or ''
    
    # Normalize unit spelling if necessary
    if unit in ['l', 'lac']:
        unit = 'lakh'
    
    # Convert based on unit
    if 'lakh' in unit:
        return value * 1e5  # 1 Lakh = 100,000
    elif 'cr' in unit or 'crore' in unit:
        return value * 1e7  # 1 Crore = 10,000,000
    else:
        return value

# Read CSV and remove ₹ before applying parsing
df["price"] = df["price"].str.replace("₹", "", regex=False)
df["price_parsed"] = df["price"].apply(parse_price)

print(df[["price", "price_parsed"]].head())
```

```python
import pandas as pd
import re

def extract_area(area_val):
    # If the value is NaN, return None
    if pd.isna(area_val):
        return None
    
    # Convert the value to a string
    s = str(area_val)
    # Search for one or more digits in the string
    match = re.search(r'(\d+)', s)
    if match:
        return int(match.group(1))
    return None

# Assuming df is your DataFrame
df['area'] = df['area'].apply(extract_area)

# Check the first few values
print(df['area'].head())
```

```python
import pandas as pd

# Convert relevant columns to numeric
df['bathroom'] = pd.to_numeric(df['bathroom'], errors='coerce')
df['numberOfRooms'] = pd.to_numeric(df['numberOfRooms'], errors='coerce')
df['balcony'] = pd.to_numeric(df['balcony'], errors='coerce')

# Create bins for price and area
df['price_bin'] = pd.cut(df['price'], bins=10)
df['area_bin']  = pd.cut(df['area'], bins=10)

# Calculate overall medians as fallback values
overall_median_bathroom = df['bathroom'].median()
overall_median_rooms    = df['numberOfRooms'].median()
overall_median_balcony  = df['balcony'].median()

# Define a function that ensures values are integers
def impute_and_convert(group, column, overall_median):
    median_value = group[column].median()
    if pd.notnull(median_value):  # If a valid median is found, round and convert
        return group[column].fillna(int(round(median_value)))
    else:  # If no valid median, use overall median
        return group[column].fillna(int(round(overall_median)))

# Apply the function to each group
df['bathroom'] = df.groupby(['price_bin', 'area_bin'], group_keys=False).apply(impute_and_convert, column='bathroom', overall_median=overall_median_bathroom)
df['numberOfRooms'] = df.groupby(['price_bin', 'area_bin'], group_keys=False).apply(impute_and_convert, column='numberOfRooms', overall_median=overall_median_rooms)
df['balcony'] = df.groupby(['price_bin', 'area_bin'], group_keys=False).apply(impute_and_convert, column='balcony', overall_median=overall_median_balcony)

# Convert to Int64 (nullable integer type)
df['bathroom'] = df['bathroom'].astype("Int64")
df['numberOfRooms'] = df['numberOfRooms'].astype("Int64")
df['balcony'] = df['balcony'].astype("Int64")

# Drop the helper binning columns
df.drop(columns=['price_bin', 'area_bin'], inplace=True)

# Verify the results
print("Missing values after imputation:")
print(df[['bathroom', 'numberOfRooms', 'balcony']].isna().sum())
```

```python
import pandas as pd

# Step 1: Remove duplicates in both files
print("Removing duplicates...")

# Load final_data.csv and remove duplicates
final_data = pd.read_csv("final_data.csv")
final_data = final_data.drop_duplicates(subset=['latitude', 'longitude'])

# Load merged_output.csv and remove duplicates
merged_output = pd.read_csv("merged_output.csv")
merged_output = merged_output.drop_duplicates(subset=['latitude', 'longitude'])

# Step 2: Convert merged_output to a dictionary for fast lookup
merged_dict = merged_output.set_index(['latitude', 'longitude']).to_dict(orient='index')

# Step 3: Process final_data in chunks to avoid memory crashes
print("Merging data in chunks...")
final_data_chunks = pd.read_csv("final_data.csv", chunksize=50000)

processed_chunks = []

for chunk in final_data_chunks:
    key_columns = ['latitude', 'longitude']
    
    # Merge using dictionary lookup
    for col in merged_output.columns:
        if col not in key_columns:  # Avoid reassigning latitude/longitude
            chunk[col] = chunk.set_index(key_columns).index.map(lambda x: merged_dict.get(x, {}).get(col, None))
    
    processed_chunks.append(chunk)  # Store processed chunk

# Step 4: Combine all processed chunks and save
final_data_updated = pd.concat(processed_chunks, ignore_index=True)
final_data_updated.to_csv("final_data_updated.csv", index=False)

print("Duplicates removed, data merged successfully! Output saved as 'final_data_updated.csv'.")
```
